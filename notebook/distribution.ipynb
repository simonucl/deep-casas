```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min, norm


# Define the shape and scale parameters
shape_params = [0.5, 1, 3]
scale_param = 1

# Generate some random numbers from the Weibull distribution for each shape parameter
x = np.linspace(-2.5, 2.5, 100)
y = []

for k in shape_params:
    y.append(weibull_min.pdf(x, k, scale=scale_param))

# Plot the probability density functions (PDFs)
plt.plot(x, y[0], label='Weibull, k = 0.5, scale = 1', linestyle=':')
plt.plot(x, y[1], label='Weibull, k = 1, scale = 1', linestyle='--')
plt.plot(x, y[2], label='Weibull, k = 3, scale = 1')
# set the gaussian distribution with mean 0 and standard deviation with kaiming normal (in = 12)
# make the label as sqrt(2/|F_in|)
plt.plot(x, norm.pdf(x, 0, np.sqrt(2/12)), label='Gaussian, mean = 0, std = sqrt(2/|T_{k}|)')
# add grid lines
plt.grid()

plt.xlabel('X')
plt.ylabel('Probability Density')
# plt.title('Weibull and Gaussian Probability Density Functions')
# put the legend above the plot
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.savefig('weibull_gaussian.png', dpi=300, bbox_inches='tight')
plt.show()
```
